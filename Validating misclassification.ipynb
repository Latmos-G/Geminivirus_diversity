```python
# Import the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
sb.set(font_scale = 2)
```

```python
# Load the dataset and display the overview of first five rows

dnn1 = pd.read_csv('misclassify.csv')  # This read out the dataset
dnn1.head() 
```

```python
# Explore if either week days or weekend has discrepancy in the success of the payment transaction

plt.figure(figsize=(35, 20))  # Chosen size of the plot

# Plot the count of the transaction (both failed and succ)in the week day

count = sb.barplot(x = 'Virus accessions within Capulavirus and Begomovirus', y = 'Percentage similarity to KC699544 based on pairwise identity', 
                   hue = 'Selected virus components', data=dnn1, palette = ['red', 'blue', 'blueviolet']
                  )
plt.legend(title = 'Selected virus components', fontsize = 30)

count.set_ylabel('Percentage (%) similarity to KC699544 based on pairwise identity',weight = 'bold', size = 30)
count.set_xlabel('Virus accessions within Capulavirus and Begomovirus')
```
